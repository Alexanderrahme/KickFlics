```python
import tensorflow as tf
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
demo_model = tf.keras.models.load_model('/content/drive/MyDrive/ResNet50_V3')
```

```python
shoe_labels = ['Adidas Continental 80 Sneaker', 'Converse Distrito 2.0 Canvas Low Sneaker', 'Under Armour Ripple 2.0 Sneaker']
```

```python
shoe_path = '/content/drive/MyDrive/Demo/C1.JPG'

cv2_imshow(cv2.resize(cv2.imread(shoe_path), (600, 800)))
```

```python
shoe_image = cv2.imread(shoe_path)
shoe_image_resized = cv2.resize(shoe_image, (224, 224))
shoe_image = np.expand_dims(shoe_image_resized, axis = 0)

prediction = demo_model.predict(shoe_image)
predicted_shoe = shoe_labels[np.argmax(prediction)]

print('Predicted shoe:', predicted_shoe)
```
